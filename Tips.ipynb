```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Solving regression problems in Machine Learning- ML is a branch of artificial intelligence that enables
# computer programs to automatically learn and improve from experience.
# ML algorithms learn from datasets and then based on the patterns identified from the datasets,
# make predictions on the unseen data.

# ML algorithms can be boradly catehorised into two types: Supervised learning and Unsupervised learning algorithms

# Supevised ML algorityhms are those algorithms where the input dataset and the corresponding output or
# true prediction is available and the algorithms try to find the relationship between inputs and outputs.
# Supervised learning algorithms are further divided into two types: regression algorithms and classification
# algortihms.

# In unsupervised ML algorithms, the true labels for the output are not known. Rather, the algorithms try to 
# find the similar patterns in the data. Clustering algorithms are a typical example of unsupervised learning.
```

```python
sns.get_dataset_names()
```

```python
tips_df= sns.load_dataset("tips")

tips_df.head()
```

```python
# Dividing data into features and labels

X=tips_df.drop(['tip'],axis=1)

y=tips_df['tip']
```

```python
X.head()
```

```python
# converting categorical data to numbers

numerical= X.drop(['sex','smoker','day','time'],axis=1)

numerical.head()
```

```python
categorical= X.filter(['sex','smoker','day','time'])

categorical.head()
```

```python
cat_numerical= pd.get_dummies(categorical, drop_first=True)
cat_numerical.head()
```

```python
X=pd.concat([numerical, cat_numerical], axis=1)
X.head()
```

```python
# Dividing Data into training and test set-

# After a machine learning algortihm has been trained, it needs to e evaluated to see how well
# it performs on unseen data. Therefore, we divide the datasets in to two sets i.e. train and test sets.
# The datsaet is trained via the train set and evaluated on the test set.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.20, random_state=0)

```

```python
# Data Scaling or normalization

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train= sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Linear Regression- Linear regression is a linear model that assumes a linear relationship between
# inputs and outputs and minimises the cost of error between the predicted and actual output using 
# functions like mean absolute error

# Advantages
# Linear regression is a simple to implement and easilty interpretable algorithm
# It takes less training time to train, even for huge datasets
# Linear regression coefficients are easy to interpret

# Disadvantages
# Performance is easily affected by outlier presence
# Assumes a linear relationship between dependent and independent variables, which 
# can result in an increased error

from sklearn.linear_model import LinearRegression

lin_reg=LinearRegression()

regressor= lin_reg.fit(X_train, y_train)

y_pred= regressor.predict(X_test)
```

```python
# Once you have trained a model and have made predictions on the test set, the next step is to know
# how well your model has performed for making predictions on the unknown test set.

# Mean absolute error- MAE is calculated by taking the average of absolute error obtained
# by subtracting real values from predicted values

# Mean Squared Error- MSE is similar to MAE. However, the error for each record is squared in the case of MSAE in order
# to punish data records with a huge difference between predicted and actual values

# Root Mean Squared Error- RMSE is simply the under the root of mean squared error
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))

print('Root Mean Absolute Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# By looking at the mean absolute error, it cna be conclude that, on average there is an error of 0.70 for
# predictions which means that on average, the predicted tip values are 0.70$ more or less than the actual
# tip values

```

```python
# Random Forest Regression- is atree based algorithm that converts features into tree nodes and then uses
# entropy loss to make predictions

# Advantages
# This model works well when you have lots of missing data or an imbalanced dataset 
# With a large number of trees, you can avoid overfitting while training. Overfitting occurs when machine 
# learning models perform better on the training set but worse on the test set.
# The random forest algorithm can solve both classification and regression tasks

from sklearn.ensemble import RandomForestRegressor

rf_reg= RandomForestRegressor(random_state=42, n_estimators=500)

regressor= rf_reg.fit(X_train, y_train)

y_pred= regressor.predict(X_test)

from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))

print('Root Mean Absolute Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Classification problems are the type of preoblems where you have to predict a discrete value
# i.e. whether or not a tumor is malignant, whether the condition of a car is good,
# whether or not a student will pass an exam
```

```python
churn_df= pd.read_csv("Churn_Modelling.csv")

churn_df.head()
```

```python
# The exited column contains information regarding whether or not the customer exited the bank after six months
```

```python
churn_df=churn_df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
X=churn_df.drop(['Exited'],axis=1)

y=churn_df['Exited']
```

```python
X.head()
```

```python
numerical=X.drop(['Geography','Gender'],axis=1)

numerical.head()
```

```python
categorical=X.filter(['Geography','Gender'])
categorical.head()
```

```python
cat_numerical= pd.get_dummies(categorical, drop_first=True)
cat_numerical.head()
```

```python
X=pd.concat([numerical, cat_numerical],axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.20, random_state=0)
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Binary Classification Problem- Are those classification problems where there are only two 
# posssible values for output label.For eg whether it will rain of not

# Logistic regression- Logistic regression is a linear model, which makes classification 
# bu passing the output of linear regression through a sigmoid function.
```

```python
from sklearn.linear_model import LogisticRegression

log_clf= LogisticRegression()

classifier= log_clf.fit(X_train, y_train)

y_pred= classifier.predict(X_test)
```

```python
# Once you have trained a model and have made predictions on the test set, the next step is to
# know how well your mode has performed on the unknown test set. There are various metrics
# to evaluate a classification method such as F1 score, recall, precision, accuracy and confusion matrix

# True Negative- True negatives are those output labels that are actually false, and the
# mdoel also predicted them as false

# True Positive- True positives are those labels that are actually true and also predicted as true by the model

# False Negative- False egatives are labels that are actually true but predicted as false by ML models

# False Positive- Labels that are actually false but predicted as true by the model.

# Precision- It is obtained by dividing true positives by the sum of true positive and false positive
# Precision= tp/(tp+fp)

#Recall: Recall is calculated by dividing true positives by the sum of true positive and
# false negative. Recall= tp/(tp+fn)

# F1 measure- It is simply the harmonic mean of precision and recall

# Accuracy- It refers to the number of correctly predicted labels divided by the
# total number of observations in the dataset. Accuracy= (tp+tn)/(tp+tn+fp+fn)

# In the case of balanced datasets, i.e. when the number of labels for each class is balanced
# accuracy can be used as an evaluation metric. For imbalanced datasets, you can use the
# F1 measure as the classification metric.
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# The output shows that for 81 percent of the records in the test set, logistic regression
# correctly predicted whether or not a customer will leave the bank
```
